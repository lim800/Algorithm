# Information Systems Journal (ISJ) 2019-2020

```python
# set up options
def enable_download_headless(browser,download_dir):
    browser.command_executor._commands["send_command"] = ("POST", '/session/$sessionId/chromium/send_command')
    params = {'cmd':'Page.setDownloadBehavior', 'params': {'behavior': 'allow', 'downloadPath': download_dir}}
    browser.execute("send_command", params)
chrome_options = Options()
# chrome_options.add_argument("--headless")
chrome_options.add_argument("--window-size=1920x1080")
chrome_options.add_argument("--disable-notifications")
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--verbose')
chrome_options.add_experimental_option("prefs", {
        "download.default_directory": "local directory",
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "safebrowsing_for_trusted_sources_enabled": False,
        "safebrowsing.enabled": False
})
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-software-rasterizer')
driver = webdriver.Chrome(options=chrome_options)

# selenium
wait = WebDriverWait(driver, 20)
url = 'https://onlinelibrary-wiley-com.ezproxy.library.unlv.edu/loi/13652575'
driver.get(url)
time.sleep(1)
driver.find_element_by_id('user').send_keys("")
driver.find_element_by_id('pass').send_keys("")
driver.find_element_by_id('pass').send_keys(Keys.ENTER)
time.sleep(2)

count, sumup = 1, 0
year = '2018'
localpath = r'local directory'
download_dir = "local directory\\ISJ_{}".format(year)
enable_download_headless(driver, download_dir)

wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 'back-btn')))
driver.find_element_by_xpath("//a[@class='back-btn']").click()
time.sleep(1)
driver.find_element_by_xpath("//span[contains(text(),'2010 - 2019')]").click()
time.sleep(1)
driver.find_element_by_xpath("//span[contains(text(),'{} - Volume')]".format(year)).click()

for i in range(1,6):
    time.sleep(2)
    driver.find_element_by_xpath("//a[contains(text(),'Issue {}')]".format(i)).click()
    wait.until(EC.presence_of_all_elements_located((By.LINK_TEXT, 'PDF')))
    pdfs = driver.find_elements_by_xpath("//a[contains(text(),'Full text')]")
    time.sleep(1)
    sumup = sumup + len(pdfs)
    print("length of pdfs:", len(pdfs), " | total pdfs:", sumup)
    
    for j in range(len(pdfs)):
        time.sleep(1)
        action = ActionChains(driver)
        el = driver.find_elements_by_xpath("//a[contains(text(),'Full text')]")[j]
        time.sleep(1)
        action.move_to_element_with_offset(el, 85, 10).click().perform()
        
        time.sleep(4)
        while len(os.listdir(localpath+'\ISJ_{}\\'.format(year))) == 0:
            action = ActionChains(driver)
            time.sleep(1)
            el = driver.find_element_by_xpath('//*[@id="app-navbar"]')
            action.move_to_element_with_offset(el, 885, 30).click().perform()

            action = ActionChains(driver)
            time.sleep(2)
            el = driver.find_element_by_xpath('//*[@id="app-navbar"]')
            action.move_to_element_with_offset(el, 750, 110).click().perform()        
            time.sleep(5)
        time.sleep(2)
        if len(os.listdir(localpath+'\ISJ_{}\\'.format(year))) == 1:    
            os.replace(localpath+'\ISJ_{}\\'.format(year) + os.listdir(localpath+'\ISJ_{}\\'.format(year))[0],\
                       localpath+'\\' + '{}_{}_{}.pdf'.format('ISJ',year,count))
        count+=1
        driver.back()
        time.sleep(1)
        
    print("issue {} is done".format(i))
    driver.back()
    time.sleep(3)
    
    
driver.close()
```

# Journal of Information Technology (JIT)

```python
# set up options
def enable_download_headless(browser,download_dir):
    browser.command_executor._commands["send_command"] = ("POST", '/session/$sessionId/chromium/send_command')
    params = {'cmd':'Page.setDownloadBehavior', 'params': {'behavior': 'allow', 'downloadPath': download_dir}}
    browser.execute("send_command", params)
chrome_options = Options()
# chrome_options.add_argument("--headless")
chrome_options.add_argument("--window-size=1920x1080")
chrome_options.add_argument("--disable-notifications")
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--verbose')
chrome_options.add_experimental_option("prefs", {
        "download.default_directory": "local directory",
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "safebrowsing_for_trusted_sources_enabled": False,
        "safebrowsing.enabled": False
})
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-software-rasterizer')
driver = webdriver.Chrome(options=chrome_options)


# selenium
url = 'https://journals-sagepub-com.ezproxy.library.unlv.edu/loi/jin'
driver.get(url)
driver.set_window_size(1000, 900)
time.sleep(1)
driver.find_element_by_id('user').send_keys("")
driver.find_element_by_id('pass').send_keys("")
driver.find_element_by_id('pass').send_keys(Keys.ENTER)
time.sleep(3)


issue = 0
count, sumup = 7, 6 #default = 1, 0
year = '2018'
localpath = r'local directory'
download_dir = "local directory\\JIT_{}".format(year)
enable_download_headless(driver, download_dir)

for i in range(2,4):
    driver.find_element_by_link_text('2020').click()
    time.sleep(1)
    driver.find_element_by_link_text('2010 - 2019').click()
    time.sleep(1)
    driver.find_element_by_link_text(year).click()
    time.sleep(5)
    a = driver.find_elements_by_partial_link_text('Issue {},'.format(i))
    if len(a) == 1:  # setup for 2019
        time.sleep(2)
        driver.find_elements_by_partial_link_text('Issue {},'.format(i))[0].click()
    else:
        driver.find_element_by_link_text(year).click()
        time.sleep(2)
        driver.find_elements_by_partial_link_text('Issue {},'.format(i))[0].click()
    time.sleep(1)
    pdfs = driver.find_elements_by_class_name('hlFld-Title')
    time.sleep(2)
    sumup = sumup + len(pdfs)
    print("length of pdfs:", len(pdfs), " | total pdfs:", sumup ," | current issue {}".format(i))
    
    for j in range(len(pdfs)):
        driver.find_elements_by_class_name('hlFld-Title')[j].click()
        time.sleep(2)
        driver.find_element_by_xpath('//*[@id="rightMobileMenuLink"]/div[1]').click()
        time.sleep(2)
        driver.switch_to.window(driver.window_handles[1])
        action = ActionChains(driver)
        el = driver.find_element_by_xpath('/html/body/embed')
        time.sleep(2)
        action.move_to_element_with_offset(el, 880, 40).click().perform()
        time.sleep(4)
        os.replace(localpath+'\JIT_{}\\'.format(year) + listdir(localpath+'\JIT_{}\\'.format(year))[0],\
                   localpath+'\\' + '{}_{}_{}.pdf'.format('JIT',year,count))
        count+=1
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        
        driver.back()
        time.sleep(3)
        
    print("issue {} is done".format(i))
    time.sleep(3)
    driver.back()
    time.sleep(1)
    
driver.close()
```

```python
# set up options
def enable_download_headless(browser,download_dir):
    browser.command_executor._commands["send_command"] = ("POST", '/session/$sessionId/chromium/send_command')
    params = {'cmd':'Page.setDownloadBehavior', 'params': {'behavior': 'allow', 'downloadPath': download_dir}}
    browser.execute("send_command", params)
chrome_options = Options()
# chrome_options.add_argument("--headless")
chrome_options.add_argument("--window-size=1920x1080")
chrome_options.add_argument("--disable-notifications")
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--verbose')
chrome_options.add_experimental_option("prefs", {
        "download.default_directory": "local directory",
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "safebrowsing_for_trusted_sources_enabled": False,
        "safebrowsing.enabled": False
})
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-software-rasterizer')
driver = webdriver.Chrome(options=chrome_options)



# selenium
url = 'https://www-sciencedirect-com.ezproxy.library.unlv.edu/journal/the-journal-of-strategic-information-systems/issues'
driver.get(url)
# driver.set_window_size(1000, 1100)
time.sleep(1)
driver.find_element_by_id('user').send_keys("")
driver.find_element_by_id('pass').send_keys("")
driver.find_element_by_id('pass').send_keys(Keys.ENTER)
time.sleep(3)


count = 1
year = '2018'
localpath = r'local directory'
download_dir = "local directory\\JSIS_{}".format(year)
enable_download_headless(driver, download_dir)

for i in range(1,4):
    driver.find_element_by_xpath("//span[contains(text(),'2020 — ')]").click()
    time.sleep(1)
    driver.find_element_by_xpath("//span[contains(text(),'{} — ')]".format(year)).click()
    time.sleep(2)
    driver.find_element_by_xpath("//span[contains(text(),'Issue {}')]".format(i)).click()
    time.sleep(2)
    pdfs = driver.find_elements_by_xpath("//dd[contains(text(),'Pages ')]")
    
    for j in range(len(pdfs)):
        if driver.find_elements_by_xpath("//dd[contains(text(),'Pages ')]")[j].text[-1].isdigit():
            action = ActionChains(driver)
            el = driver.find_elements_by_xpath("//dd[contains(text(),'Pages ')]")[j]
            action.move_to_element(el).perform()
            driver.find_element_by_tag_name('body').send_keys(Keys.ARROW_DOWN)
            time.sleep(1)
            action.move_to_element_with_offset(el, 20, 50).click().perform()
            time.sleep(3)
            driver.switch_to.window(driver.window_handles[1])
            action = ActionChains(driver)
            el = driver.find_element_by_xpath('/html/body/embed')
            time.sleep(2)
            action.move_to_element_with_offset(el, 765, 30).click().perform()
            time.sleep(4)
            os.replace(localpath +'\JSIS_{}\\'.format(year) + os.listdir(localpath +'\JSIS_{}\\'.format(year))[0],\
                       localpath+'\\' + '{}_{}_{}.pdf'.format('JSIS',year,count))
            count+=1
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
    
    print("issue {} is done".format(i))
    driver.back()
    time.sleep(2)
    driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
    time.sleep(1)
            

driver.close()
```
