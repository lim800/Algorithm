# LeetCode: 1021. Remove Outermost Parentheses

```python
S = "(()())(())(()(()))"
class Solution:
    def removeOuterParentheses(self, S) -> str:
        starter = 0
        count = 0
        answer = ""
        for i in range(len(S)):
            if S[i] == "(":
                count +=1
            else:
                count -=1
            if count == 0:
                answer = answer + S[starter+1:i]
                starter = i + 1        
        return answer
    
sol = Solution()
sol.removeOuterParentheses(S)
```

# LeetCode: 1450. Number of Students Doing Homework at a Given Time

```python
startTime = [4]
endTime = [4]
queryTime = 4

class Solution:
    def busyStudent(self, startTime, endTime, queryTime) -> int:
        cnt = 0
        for i in range(len(startTime)):
            if queryTime in range(startTime[i], endTime[i]+1):
                cnt += 1
        return cnt
        
sol = Solution()
sol.busyStudent(startTime, endTime, queryTime)
```

# LeetCode: 1475. Final Prices With a Special Discount in a Shop

```python
# prices = [8,4,6,2,3]
prices = [10,1,1,6]

class Solution:
    def finalPrices(self, prices):
        final = []
        for i in range(len(prices)):
            pointer = 1
            final.append("N")
            while i+pointer<len(prices):
                if prices[i] >= prices[i+pointer]:
                    final[i] = (prices[i]-prices[i+pointer])
                    break
                pointer += 1
            if final[i]=="N":   
                final[i] = prices[i]        
        return final
    
    
sol = Solution()
sol.finalPrices(prices)
```

# LeetCode: 832. Flipping an Image

```python
A = [[1,1,0],[1,0,1],[0,0,0]]

class Solution:
    def flipAndInvertImage(self,A):
        for i in A:
            front, back = 0, (len(i)-1)
            while True:
                if len(i)%2!=0 and front == back:
                    print("front: ", front, " back: ", back, "value: ", i[front])
                    i[front] = self.flip(i[front])
                    break
                    
                if i[front] == i[back]:
                    i[front], i[back] = self.flip(i[front]), self.flip(i[back])
                
                if len(i)%2==0 and front +1 == back:
                    break
                front +=1
                back -=1
        return A        
        
    def flip(self,num):
        if num == 1:
            return 0
        else:
            return 1
        
sol = Solution()        
sol.flipAndInvertImage(A)
```

# 프로그래머스 스킬트리

<제한 조건>
    - 스킬은 알파벳 대문자로 표기하며, 모든 문자열은 알파벳 대문자로만 이루어져 있습니다.
    - 스킬 순서와 스킬트리는 문자열로 표기합니다.
    - 예를 들어, C → B → D 라면 CBD로 표기합니다
    - 선행 스킬 순서 skill의 길이는 2 이상 26 이하이며, 스킬은 중복해 주어지지 않습니다.
    - skill_trees는 길이 1 이상 20 이하인 배열입니다.
    - skill_trees의 원소는 스킬을 나타내는 문자열입니다.
    - skill_trees의 원소는 길이가 2 이상 26 이하인 문자열이며, 스킬이 중복해 주어지지 않습니다.

```python
#제출용
import re
def solution(skill, skill_trees):
    tree = skill_trees.copy()
    for i in list(skill):
        tree = (list(filter(lambda x : not i in x, tree)))
    cnt= len(tree)
    for j in tree:
        skill_trees.remove(j)
    for k in skill_trees:
        sen = re.sub(r"[^{}]".format(skill), "", k)
        for l, m in enumerate(list(skill)):
            sen = sen.replace(m,str(l))
        aaa = list(filter(lambda x: x != str(list(sen).index(x)), list(sen)))
        if not aaa:
            cnt+=1
    answer = cnt
    return answer
```

# LeetCode 999. Available Captures for Rook

```python
class Solution:
    def numRookCaptures(input) -> int:
        count = 0
        col = -1
        row = -1
        dic = {'N':0,'S':0,'E':0,'W':0}
        add = 1
        deduct = 1
        for i in range(0,len(input)):
            if 'R' in input[i]:
                row = i
                col = input[i].index('R')
        while(True):
            if dic['N'] == 0 and (row - deduct)>=0:
                if input[row - deduct][col] != '.':
                    if input[row - deduct][col] == 'p':
                        count +=1
                    dic['N'] =1
            else:
                dic['N'] = 1

            if dic['S'] == 0 and (row + add)<=len(input)-1:
                if input[row + add][col] != '.':
                    if input[row + add][col] == 'p':
                        count +=1
                    dic['S'] =1
            else:
                dic['S'] = 1

            if dic['E'] == 0 and (col - deduct)>=0:
                if input[row][col - deduct] != '.':
                    if input[row][col - deduct] == 'p':
                        count +=1
                    dic['E'] =1
            else:
                dic['E'] = 1

            if dic['W'] == 0 and (col + add)<=len(input[col])-1:
                if input[row][col + add] != '.':
                    if input[row][col + add] == 'p':
                        count +=1
                    dic['W'] =1
            else:
                dic['W'] = 1            

            deduct +=1
            add +=1    
            if dic['N'] == 1 and dic['S'] == 1 and dic['E'] == 1 and dic['W'] == 1:
                break       
        
        
        return count
```

# LeetCode 896. Monotonic Array

```python
A = [1,1,1,1,1,1,1]

class Solution:
    def isMonotonic(A) -> bool:
        if len(A) == 1:
            return True
        direction = 1
        for j in range(len(A)-1):
            if A[j] != A[j+1]:
                if A[j] >= A[j+1]:
                    direction = -1
                    break
                break
        for i in range(len(A)-1):
            if A[i]*direction > A[i+1]*direction:
                return False
        return True
         
Solution.isMonotonic(A)        
        
        
```

# LeetCode 812. Largest Triangle Area

```python
points =  [[2,5],[7,7],[10,8],[10,10],[1,1]]
```

```python
class Solution:
    def largestTriangleArea(points):
        ans = []
        for A in points:
            for B in points:
                if B != A:
                    for C in points:
                        if C !=A and C !=B:
                            ans.append(abs((A[0]*(B[1]-C[1]) + B[0]*(C[1]-A[1]) + C[0]*(A[1]-B[1]))/2))
        return max(ans)
        
```
